```python
import requests
import pandas as pd
import plotly.express as px
from datetime import datetime as dt
```

```python
header = {'Authorization': f'Bearer {API_KEY}'}
```

```python
# stocksTicker, multiplier, timespan, from, to, adjusted, sort, limit

params = {
    'stocksTicker': 'GOGO',
    'multiplier': 1,
    'timespan': 'day',
    'from': '2022-01-01',
    'to': dt.now().strftime('%Y-%m-%d'),
    }
print(params)
```

```python
endpoint = '{domain}/v2/aggs/ticker/{stocksTicker}/range/{multiplier}/{timespan}/{start}/{end}'
```

```python
#/v2/aggs/ticker/{stocksTicker}/range/{multiplier}/{timespan}/{from}/{to}
url = endpoint.format(
    domain=DOMAIN,
    stocksTicker=params['stocksTicker'],
    multiplier=params['multiplier'],
    timespan=params['timespan'],
    start=params['from'],
    end=params['to']
)

print(url)
```

```python
rsp = requests.get(url, headers=header)
```

```python
data = rsp.json()
data
```

```python
data.keys()
```

```python
results = data['results']
results
```

```python
close_price = [var['c'] for var in results]
```

```python
time = [var['t'] for var in results]
```

```python
print(close_price)
print(time)
```
