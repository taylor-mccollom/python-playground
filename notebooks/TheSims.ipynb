```python
import random
import time
```

```python
class Disease:
    
    def __init__(self, name, symptoms, fatal, severity, min_age_onset):
        # string
        self.name = name
        # list of strings
        self.symptoms = symptoms
        # boolean
        self.fatal = fatal
        # scale of 1-10
        self.severity = severity
        #TODO exclude minimum_age_of_onset until later
        self.min_age_onset = min_age_onset
        
    @property
    def prognosis(self):
        return f'Congratulations...you have {self.name} with symptoms: {self.symptoms}'
```

```python
class Career:
    
    def __init__(self, name, salary, industry):
        # TODO exclude social_level until later
        self.name = name
        self.salary = salary
        self.industry = industry
    # TODO: can we duplicate less code across these two methods?   
    def promotion(self):
        pct_increase = random.randint(0,1)
        self.salary += self.salary*pct_increase
        
    def demotion(self):
        pct_decrease = random.randint(0,1)
        self.salary -= self.salary*pct_decrease
```

```python
class Environment:
    
    def __init__(self, civilization_level, possible_careers, min_working_age):
    # civilization_level is on a scale 1-3 where 1-3 correspond to 1st, 2nd,
    # and 3rd world countries
        self.civilization_level = civilization_level
        self.possible_careers = possible_careers
        self.min_working_age = min_working_age
```

```python
class Sim:
    
    def __init__(self, 
                first_name, 
                last_name, 
                height, 
                weight, 
                charisma, 
                environment, 
                intelligence,
                dexterity,
                possible_diseases):
        self.first_name = first_name
        self.last_name = last_name
        self.height = height
        self.weight = weight
        # sim starts as a baby
        self.age = 0
        # charisma is on scale 1-10 and static different meaning relative
        # to age
        self.charisma = charisma
        # environment is going to be an Environment object
        self.environment = environment
        # intelligence is going to be on scale 1-10
        self.intelligence = intelligence
        # dexterity on a scale from 1-10, static unless extra time
            # then can increase with age
        self.dexterity = dexterity
        # life_expectancy random between 5-120
        self.life_expectancy = random.randint(5,120)
        # possible diseases is a list of disease objects
        self._possible_diseases = possible_diseases
        self.diseases = []
        self.is_alive = True
        self.career = None
        
    @property
    def full_name(self):
        return f'{self.first_name} {self.last_name}'
    
    def birthday(self):
        self.age += 1
        print(f'Congratulations you turned {self.age}!')
        
        got_disease = random.choices([True, False], weights=[0.17, 0.83])[0]
        if got_disease:
            disease = random.choice(self._possible_diseases)
            if disease not in self.diseases and self.age >= disease.min_age_onset:
                print('You got a disease.')
                print(disease.prognosis)
                self.diseases.append(disease)
        
        if self.age >= self.environment.min_working_age and self.career is None:
            self.career = random.choice(self.environment.possible_careers)
            print('You got a job.')
            print(f'You are {self.career.name}.')
            print(f'You work in the {self.career.industry} industry.')
            print(f'Your salary is {self.career.salary}.')
        
        if self.age == self.life_expectancy:
            print('You Died.')
            self.is_alive = False
            
```

```python
rabies = Disease(name='Rabies', symptoms=['mouth foam'], fatal=True, severity=8, min_age_onset=13)
ebola = Disease(name='Ebola', symptoms=['bleeding orifices'], fatal=True, severity=10, min_age_onset=26)
black_plague = Disease(name='Black Plague', symptoms=['dysentary'], fatal=False, severity=5, min_age_onset=22)

possible_diseases = [rabies, ebola, black_plague]
```

```python
engineer = Career(name='Engineer', salary=160_000, industry='civil')
laborer = Career(name='Laborer', salary=20_000, industry='civil')
child_fact_worker = Career(name='Child Factory Worker', salary=10_000, industry='civil')
child_soldier = Career(name='Child Soldier', salary=1_000, industry='civil')
subversive = Career(name='Subversive', salary=180_000, industry='civil')
doctor = Career(name='Doctor', salary=500_000, industry='civil')

first_world_careers = [engineer, doctor, laborer]
second_world_careers = [child_fact_worker, subversive, doctor, laborer]
third_world_careers = [child_fact_worker, subversive, child_soldier, doctor, laborer]
```

```python
first_world = Environment(civilization_level=1, possible_careers=first_world_careers, min_working_age=16)
second_world = Environment(civilization_level=2, possible_careers=second_world_careers, min_working_age=11)
third_world = Environment(civilization_level=3, possible_careers=third_world_careers, min_working_age=3)
```

```python
sim1 = Sim(first_name='Eugene', 
        last_name='Berkowitz', 
        height=3, 
        weight=11, 
        charisma=7, 
        environment=first_world, 
        intelligence=6, 
        dexterity=10,
        possible_diseases=possible_diseases)
```

```python
while sim1.is_alive:
    time.sleep(0.0)
    sim1.birthday()
```
